```python
lines = updatedpoems.split("\n\n")
lines
```

```python
titles = []
nontitles = []

for l in lines: 
    if l.startswith("\n") == True:
        titles.append(l)
    else:
        nontitles.append(l)

titles = [l[1:] for l in titles]
```

Looking at nontitles, this would be tough to parse through since the poems are different line lengths. Let's go back to lines and see if there is a way to parse through between lines that are present in titles. Thinking through this, I think it would be better to try again by separating the poems first. Based on what came out of the others, it looks like each poem is separated by 3 line breaks, not 2.

```python
test = updatedpoems.split("\n\n\n")
test[0]
```

```python
test_titles = []
test_lines = []

for t in test:
    a = t.split("\n\n")
    print(a[:3])
    test_titles.append(a[0])
    test_lines.append(a[1:])
```

```python
test_titles[0] = "Introduction"
```

```python
test_titles
```

```python
test_lines
```

```python
words = []

for w in test_lines:
    a = " ".join(w)
    b = a.replace("\n", " ")
    c = b.split(" ")
    words.append(c)
```

```python
words[1:3]
```

```python
print(len(test_titles))
print(len(test_lines))
print(len(words))
```

Yay!! It worked!!!

```python
poems_df = pd.DataFrame(test_titles)
```

```python
poems_df["Lines"] = test_lines
poems_df["Words"] = words
```

```python
poems_df.columns = ["Title", "Lines", "Words"]
```
